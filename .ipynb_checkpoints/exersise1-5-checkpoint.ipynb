```python
import numpy as np
import matplotlib.pyplot as plt 
#plt.plot([1,2,3,4])
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X=np.arange(-50,50,0.5)
Y=np.arange(-50,50,0.5)
X,Y = np.meshgrid(X,Y)
Z=2*X+5*Y
ax.plot_wireframe(X,Y,Z)
ax.scatter(0,0,0,"z",50,"green")
#ax.plot_wireframe(X,Y,Z+1111)
Z=2*X+5*Y+500
ax.plot_wireframe(X,Y,Z)
show()
```

```python
fig = figure()
ax = Axes3D(fig)
X=np.arange(-50,50,0.5)
Y=np.arange(-50,50,0.5)
X,Y = np.meshgrid(X,Y)
Z=2*X**2+5*Y**2
ax.plot_wireframe(X,Y,Z)
ax.plot_wireframe(X,Y,-Z)
show()
```
