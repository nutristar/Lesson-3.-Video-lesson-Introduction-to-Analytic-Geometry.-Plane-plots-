```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x=np.linspace(-4,4,3000)
y=3*x**2+1 #parabola
y2=(16-x**2)**(1/2)

plt.plot(x,y)
plt.plot(x,y2)
#plt.plot(x,y3)

plt.xlabel("x")
plt.ylabel("y")
```

```python
x=np.linspace(-4,4,10000)
y2=(16-x**2)**(1/2)
plt.plot(x,y2)
y3=-(16-x**2)**(1/2)        #окружность
plt.plot(x,y3)

plt.xlabel("x")
plt.ylabel("y")
```

```python
x=np.linspace(-4,4,10000)
y2=(16-2*(x**2))**(1/2)
plt.plot(x,y2)
y3=-(16-2*(x**2))**(1/2)       #элипс
plt.plot(x,y3)

plt.xlabel("x")
plt.ylabel("y")
```

```python
x=np.linspace(-40,40,10000)
y2=(9*x**2-9)/16
plt.plot(x,y2)
y3=-(9*x**2-9)/16       #hiperbola
plt.plot(x,y3)

plt.xlabel("x")
plt.ylabel("y")
```
