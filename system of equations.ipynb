```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
from scipy.optimize import fsolve
def f(x):
    return (np.sin(x+1))
x0=fsolve(f,1)
print(x0)

x=np.linspace(-3*np.pi, 3*np.pi, 201)
plt.plot(x,f(x))
plt.plot([-10,10],[0,0],color="red", linewidth=0.75, linestyle = "--")
plt.grid(True)
plt.show()
```

```python
x=np.linspace(-2,3,201)
plt.plot(x,(np.exp(x)-1+x)/x)
plt.plot(x,x**2-1)
```

```python
def equations(p):
    x,y = p
    return (x**2-1-y, (np.exp(x)+x*(1-y)-1)>0)
x1,y1 = fsolve(equations,(-2,2))
print(x1,y1)
x2,y2 = fsolve(equations,(2,4))
print(x2,y2)
```
